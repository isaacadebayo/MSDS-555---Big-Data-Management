*********************************************************************
# Checking for null, note ? is used as a descriptor for unknown
*********************************************************************

```python
from pyspark.sql.functions import col
from functools import reduce

ui_port = "4050"
spark = SparkSession.builder \
    .appName("SecureSparkSession") \
    .config("spark.ui.port", ui_port) \
    .config("spark.driver.extraJavaOptions", "-Djava.security.manager=allow") \
    .getOrCreate()

df = spark.read.csv("adult.csv", header=True, inferSchema=True)

columns_to_check = ["educational-num", "age", "workclass", "fnlwgt",  "education", "marital-status", "occupation", "relationship", "race", "gender", "capital-gain", "capital-loss", "hours-per-week","native-country", "income"]

# Create a combined condition for any of the columns being null
null_conditions = [col(c).isNull() for c in columns_to_check]
combined_null_condition = reduce(lambda a, b: a | b, null_conditions)

is_null = df.filter(combined_null_condition)
is_null.show()
```
