*********************************************************************
# Using the order by function to get marital status
*********************************************************************

```python
ui_port = "4045"
spark = SparkSession.builder \
    .appName("SecureSparkSession") \
    .config("spark.ui.port", ui_port) \
    .config("spark.driver.extraJavaOptions", "-Djava.security.manager=allow") \
    .getOrCreate()

df = spark.read.csv("adult.csv", header=True, inferSchema=True)
salary = df.orderBy("marital-status")
salary.show(20)
```

*********************************************************************
# Filtering Age greater than 40
*********************************************************************

```python
ui_port = "4045"
spark = SparkSession.builder \
    .appName("SecureSparkSession") \
    .config("spark.ui.port", ui_port) \
    .config("spark.driver.extraJavaOptions", "-Djava.security.manager=allow") \
    .getOrCreate()

df = spark.read.csv("adult.csv", header=True, inferSchema=True)
df.filter(col("Age") < 40).count()
```

*********************************************************************
# Using the groupby function to aggregate education attainment
*********************************************************************

```python
ui_port = "4050"
spark = SparkSession.builder \
    .appName("SecureSparkSession") \
    .config("spark.ui.port", ui_port) \
    .config("spark.driver.extraJavaOptions", "-Djava.security.manager=allow") \
    .getOrCreate()

df = spark.read.csv("adult.csv", header=True, inferSchema=True)
education = df.groupBy("Education").count().orderBy(col("count").asc())
education.show()
```

*********************************************************************
# Using the select function to get distinct row (occupation)
*********************************************************************

```python
ui_port = "4050"
spark = SparkSession.builder \
    .appName("SecureSparkSession") \
    .config("spark.ui.port", ui_port) \
    .config("spark.driver.extraJavaOptions", "-Djava.security.manager=allow") \
    .getOrCreate()

df = spark.read.csv("adult.csv", header=True, inferSchema=True)
distinct_row = df.select("occupation").distinct()
distinct_row.show()
```

*********************************************************************
# Using the rename method to rename column
*********************************************************************

```python
ui_port = "4050"
spark = SparkSession.builder \
    .appName("SecureSparkSession") \
    .config("spark.ui.port", ui_port) \
    .config("spark.driver.extraJavaOptions", "-Djava.security.manager=allow") \
    .getOrCreate()

df = spark.read.csv("adult.csv", header=True, inferSchema=True)
rename_column = df.withColumnRenamed("educational-num", "educational-years")
rename_column.show()
```

*********************************************************************
# Adding new calculated column to the dataframe
*********************************************************************

```python
ui_port = "4050"
spark = SparkSession.builder \
    .appName("SecureSparkSession") \
    .config("spark.ui.port", ui_port) \
    .config("spark.driver.extraJavaOptions", "-Djava.security.manager=allow") \
    .getOrCreate()

df = spark.read.csv("adult.csv", header=True, inferSchema=True)
# Random column
new_column = df.withColumn("end-of-education", df.age - df["educational-num"])
new_column.show()
```
